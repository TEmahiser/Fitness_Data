```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#import my workout data csv called "Workout_Data"
df = pd.read_csv('workout_data.csv')

df.head()
```

In this notebook I'm going to ask certain questions of the data set. 

First, how many unique days exist in the data set? Said another way, how much time does this dataset encompass? From the previous cell, the date field is a string that includes the time stamp as well. I need to separate those 2 pieces of info.

```python
#This identifies the data type of the date column. 

print(df["Date"].dtype)

#creates a shortened version of the date string, only including the mmddyyy portion.
df['short_Date'] = df['Date'].str.slice(0,10)

df.head()
```

```python
#unique dates is calculated by using nunique() which is a count of all the unique values in a data column. IN this case it's 
# date field.

unique_dates = df["short_Date"].nunique()

print(unique_dates)
```

This shows that I have 218 days of workout data in this dataset, which means that I was working out for 218 days. If I was working out everyother day, I could assume that I was working out around 4 days a week so I could divide this number by 4 and get the number of weeks (and then months) this data set includes.

```python
numberofWeeks = unique_dates/4

print(numberofWeeks)
```

According to this number, this workout data includes over a full year of data. To confirm this, I can isolate the year in the date field and count how many unique values are there.

```python
df["year"] = df['short_Date'].str.slice(0,4)


unique_Year = df["year"].nunique()
print(df["year"])
print(unique_Year)
```

This makes sense, starting in 2019 and going all the way through 2020 and eventually exporting the data sometime in 2021.
